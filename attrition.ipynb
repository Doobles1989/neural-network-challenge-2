## Part 1: Preprocessing

```python
# Install tensorflow package
%pip install tensorflow

# Import our dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from keras.layers import Input, Dense

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition", "Department"]]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
# Select 10 columns for features
selected_columns = [
    'Age', 'DistanceFromHome', 'HourlyRate', 'YearsWithCurrManager',
    'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears',
    'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion'
]


# Create X_df using your selected columns
X_df = attrition_df[selected_columns]


# Show the data types for X_df
X_df.dtypes

```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=78)

```

```python
# Convert your X data to numeric data types however you see fit
X_train = X_train.apply(pd.to_numeric, errors='coerce')
X_test = X_test.apply(pd.to_numeric, errors='coerce')

# Add new code cells as necessary
X_test.fillna(0, inplace=True)


(X_train.head())
(X_test.head())
(y_train.head())
(y_test.head())
```

```python
# Create a StandardScaler
scaler = StandardScaler()


# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)


# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

print(X_train_scaled.shape)
print(X_test_scaled.shape)
print(y_train.shape)
print(y_test.shape)
print(X_train_scaled)
print(X_test_scaled)
print(y_train)
print(y_test)


```

```python
# Create a OneHotEncoder for the Department column
enc_department = OneHotEncoder(sparse_output=False)
department_encoded = enc_department.fit_transform(attrition_df[['Department']])
department_encoded_df = pd.DataFrame(department_encoded, columns=enc_department.get_feature_names_out(['Department']))

# Fit the encoder to the training data
enc_department.fit(y_train[['Department']])
enc_department.fit(y_test[['Department']])


# Create two new variables by applying the encoder to the training and testing data])
y_train_encoded = enc_department.transform(y_train[['Department']])
y_test_encoded = enc_department.transform(y_test[['Department']])
(y_train_encoded)
```

```python
# Create a OneHotEncoder for the Attrition column
enc_attrition = OneHotEncoder(sparse_output=False)
attrition_encoded = enc_attrition.fit_transform(attrition_df[['Attrition']])
attrition_encoded_df = pd.DataFrame(attrition_encoded, columns=enc_attrition.get_feature_names_out(['Attrition']))

# Fit the encoder to the training data
enc_attrition.fit(y_train[['Attrition']])
enc_attrition.fit(y_test[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded = enc_attrition.transform(y_train[['Attrition']])
y_test_encoded = enc_attrition.transform(y_test[['Attrition']])

(y_train_encoded)

```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_features = X_train_scaled.shape[1]


# Create the input layer
input_layer = Input(shape=(10,))

# Create at least two shared layers
shared1 = Dense(64, activation='relu', name='shared1')(input_layer)
shared2 = Dense(128, activation='relu', name='shared2')(shared1)

(shared1)
(shared2)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_branch = shared1
department_branch = shared2

# Create the hidden layer
department_hidden = layers.Dense(32, activation='relu', name='department_hidden')(department_branch)

# Create the output layer
department_output = layers.Dense(3, activation='softmax', name='department_output')(department_hidden)

(department_output)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attrition_branch = shared1
attrition_branch = shared2

# Create the hidden layer
attrition_branch = layers.Dense(units=8, activation='relu')(attrition_branch)

# Create the output layer
attrition_output = layers.Dense(units=2, activation='softmax', name='attrition_output')(attrition_branch)

(attrition_output)

```

```python
# Create the model
model = Model(
    inputs=input_layer,
    outputs=[department_output, attrition_output]
)

# Compile the model
model.compile(
    optimizer='adam',
    loss={
        'department_output': 'categorical_crossentropy',
        'attrition_output': 'binary_crossentropy'
    },
    metrics=['accuracy']
)

# Summarize the model
model.summary()
```

```python
(y_train.columns)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Encode the Department column for y_train and y_test
y_train_department_encoded = enc_department.transform(y_train[['Department']])
y_test_department_encoded = enc_department.transform(y_test[['Department']])

# Encode the Attrition column for y_train and y_test
y_train_attrition_encoded = enc_attrition.transform(y_train[['Attrition']])
y_test_attrition_encoded = enc_attrition.transform(y_test[['Attrition']])

# Train the model
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

model.fit(X_train_scaled, {'department_output': y_train_department_encoded, 'attrition_output': y_train_attrition_encoded},
          epochs=100,
          batch_size=32,
          validation_split=0.2,  # Use a portion of the training data for validation
          shuffle=True,
          verbose=2,
          callbacks=[early_stopping])
```

```python
# Evaluate the model with the testing data
results = model.evaluate(X_test_scaled, {'department_output': y_test_department_encoded, 'attrition_output': y_test_attrition_encoded}, verbose=2)
model_loss = results[0]
department_loss = results[1]
attrition_loss = results[2]
department_accuracy = results[3]
attrition_accuracy = results[4]
(f"Total Loss: {model_loss}, Department Loss: {department_loss}, Attrition Loss: {attrition_loss}")
(f"Department Accuracy: {department_accuracy}, Attrition Accuracy: {attrition_accuracy}")
```

1. No, accuracy may not be the best metric due to potential class imbalance in the dataset. If one class significantly outnumbers another, a model could achieve high accuracy by predominantly predicting the majority class, which would not reflect its true performance. Metrics such as precision, recall, or F1-score provide a more nuanced understanding of the model's effectiveness, particularly in identifying the minority class, which is often more critical in applications like attrition prediction.

2. I used the softmax activation function for both output layers. Softmax is appropriate for multi-class classification tasks, such as predicting the Department, as it converts the raw output scores into probabilities that sum to one, allowing for easy interpretation of class likelihoods. For the Attrition output, while softmax can be used, a sigmoid activation function is typically preferred for binary classification. However, in this case, softmax was chosen to maintain consistency across both outputs since we are using categorical cross-entropy for loss calculation.

3. Several strategies could enhance the model's performance:
Hyperparameter Tuning: Experimenting with different architectures, such as varying the number of layers, units per layer, and learning rates, could lead to better results.
Regularization Techniques: Implementing dropout layers or L2 regularization can help prevent overfitting, especially if the model is complex relative to the training data size.
Feature Engineering: Creating new features or transforming existing ones (e.g., polynomial features, interaction terms) could provide the model with more informative inputs.
Ensemble Methods: Combining predictions from multiple models (e.g., bagging, boosting) can improve overall performance and robustness.
Class Weighting: If the classes are imbalanced, applying class weights during training can help the model focus more on the minority class, improving its predictive capability for that group.
